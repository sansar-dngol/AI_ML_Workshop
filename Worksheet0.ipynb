2.Ask the user to input the value to be converted.

3.Perform the conversion and display the result.

4.Handle potential errors, such as invalid input or unsupported conversion types. • Requirements:

5.Functions: Define at least one function to perform the conversion.

6.Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).

7.User Input: Prompt the user to select the conversion type and input the value.

8.Docstrings: Include a docstring in your function to describe its purpose, parameters, and return value. • Conversion Options:

9.Length: – Convert meters (m) to feet (ft). – Convert feet (ft) to meters (m).

10.Weight: – Convert kilograms (kg) to pounds (lbs). – Convert pounds (lbs) to kilograms (kg).

11.Volume: – Convert liters (L) to gallons (gal). – Convert gallons (gal) to liters (L)

```python
def convert_units(conversion_type, value):
    conversions = {
        "m_to_ft": value * 3.28084,
        "ft_to_m": value / 3.28084,
        "kg_to_lbs": value * 2.20462,
        "lbs_to_kg": value / 2.20462,
        "l_to_gal": value * 0.264172,
        "gal_to_l": value / 0.264172
    }

    if conversion_type in conversions:
        return conversions[conversion_type]
    else:
        raise ValueError("Unsupported conversion type.")

def unit_conversion_program():
    print("Select conversion type:")
    print("1. m_to_ft")
    print("2. ft_to_m")
    print("3. kg_to_lbs")
    print("4. lbs_to_kg")
    print("5. l_to_gal")
    print("6. gal_to_l")

    choice_map = {
        "1": "m_to_ft",
        "2": "ft_to_m",
        "3": "kg_to_lbs",
        "4": "lbs_to_kg",
        "5": "l_to_gal",
        "6": "gal_to_l"
    }

    try:
        choice = input("Enter choice (1-6): ")
        if choice not in choice_map:
            raise ValueError("Invalid choice.")

        value = float(input("Enter value to convert: "))
        result = convert_units(choice_map[choice], value)

        print(f"Converted value: {result}")

    except ValueError as e:
        print("Error:", e)

unit_conversion_program()
```

Create a Python program that performs various mathematical operations on a list of numbers.

```python
def calculate_sum(numbers):
    return sum(numbers)

def calculate_average(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return max(numbers)

def find_minimum(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return min(numbers)

def math_operations_program():
    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter choice (1-4): ")
        numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

        if not numbers:
            raise ValueError("List cannot be empty.")

        if choice == "1":
            print("Result:", calculate_sum(numbers))
        elif choice == "2":
            print("Result:", calculate_average(numbers))
        elif choice == "3":
            print("Result:", find_maximum(numbers))
        elif choice == "4":
            print("Result:", find_minimum(numbers))
        else:
            print("Invalid choice.")

    except ValueError:
        print("Error: Please enter valid numeric input.")

math_operations_program()
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.
• Requirements:
– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.
– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(item)
        else:
            result.append(item)
    return result

def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]
    return element

def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

def average_nested(lst):
    flat_list = deep_flatten(lst)
    if len(flat_list) == 0:
        raise ValueError("List is empty.")
    return sum(flat_list) / len(flat_list)

nested = [1, 2, [3, 4, [5, 5, 6]], [7, [8, 9]], 10, [11, [12]], 5]

flat_one = flatten(nested)
print("flatten (one level):", flat_one)

val = access_nested_element(nested, [2, 2, 1])
print("access_nested_element([2,2,1]):", val)

total = sum_nested(nested)
print("sum_nested:", total)

cleaned = remove_element(nested, 5)
print("remove_element 5:", cleaned)

mx = find_max(nested)
print("find_max:", mx)

five_count = count_occurrences(nested, 5)
print("count_occurrences of 5:", five_count)

completely_flat = deep_flatten(nested)
print("deep_flatten:", completely_flat)

avg = average_nested(nested)
print("average_nested:", avg)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
print("A * A_inv:\n", np.dot(A, A_inv))

# 2. Prove AB ≠ BA
print("AB:\n", np.dot(A,B))
print("BA:\n", np.dot(B,A))

# 3. Prove (AB)^T = B^T A^T
left = np.dot(A,B).T
right = np.dot(B.T, A.T)
print("(AB)^T:\n", left)
print("B^T A^T:\n", right)
```
